This notebook analyzes drug data from the National Survey on Drug Use and Health (NSDUH). This data is easily accesible by viewing the [CORGIS collection](https://think.cs.vt.edu/corgis/json/drugs/drugs.html). The data contains information about drug use across different states and the District of Columbia. It also has records of drug usage rates among different age groups from the years 2002 to 2014. This data is important because it highlights the amount of people affected by drugs in the United States.

# Loading the Data

```python
# Load
import requests
reports = requests.get("https://think.cs.vt.edu/corgis/json/drugs/drugs.json").json()
# Preview the data
from pprint import pprint
pprint(reports[0])
```

# Histogram Analysis

```python
# Preprocess

"""This for loop appends a list with the number of young adults who smoked in the past 
month for each state and the District of Columbia. All the data extracted was from the year 2014."""

Tobacco_rates_for_2014 = []
for report in reports:    
    # Filter out other years besides 2014
    if report["Year"] == 2014:
        Tobacco_usage = report["Rates"]["Tobacco"]["Cigarette Past Month"]["18-25"]
        Tobacco_rates_for_2014.append(Tobacco_usage)
import matplotlib.pyplot as plt
plt.hist(Tobacco_rates_for_2014)
plt.title("Tobacco Rates Across the Country in 2014 for Young Adults")
plt.xlabel("Percentage of 18 to 25 year olds Who Smoke")
plt.ylabel("Number of States")
plt.show()
```

This histogram shows the percentage of people around the age 18-25 in each state who smoke. The histogram shows that most states have around the same amount of percentage of young adults who smoke. According to the histogram, the mean percentage of young adults who smoke for each state is about 30%.
